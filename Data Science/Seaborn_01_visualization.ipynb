### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
#scatter plot
tips=sns.load_dataset('tips')
tips.head()
```

```python
#scatter plot -> axis level function
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker')
```

```python
# fig level function -> squre shape
#sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter')
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',hue='sex',style='smoker',size='size')
```

```python
#line plot-> fig level finction
gap=px.data.gapminder()
gap.head()
```

```python
temp=gap[gap['country']=='India']
```

```python
sns.lineplot(data=temp,x='year',y='lifeExp')
```

```python
#using relplot
sns.relplot(x='year',y='lifeExp',data=temp,kind='line')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent', size='continent')
```

```python
#facet plot
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent', size='continent',col='country')
```

```python
#more Example of facet plot
# it will work with fig level func
sns.relplot(kind='scatter', data=tips, x='total_bill', y='tip', col='time', row='smoker')
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
#figure lvel ->displot
#axes lvel-> histplot -> kdeplot -> rugplot
```

```python
#Plot univariate Histogram
sns.histplot(data=tips,x='total_bill')
```

```python
#plotting univariate histogram fig-lvel
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
#bins distribution
sns.displot(data=tips,x='total_bill',kind='hist',bins=2)
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(data=tips, x='day', kind='hist')
```

```python
#hue parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex')
#
```

```python
#element para
sns.displot(data=tips, x='tip', kind='hist',hue='sex',element='step')
#
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
sns.displot(data=titanic,x='age',kind='hist',hue='sex',multiple='stack')
```

```python
#facet -> only on distplot
sns.displot(data=titanic,x='age',kind='hist',hue='sex',multiple='stack',col='class',col_wrap=2)
#
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
#hue ->fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
#bivarant kde plot
sns.displot(data=tips, x='total_bill', y='tip',kind='kde')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
#heatmap ->axis level function
#plot rectangular data as a colour-encoded data
df=gap.pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(df,cmap='Blues')
plt.figure(figsize=(15,10))
#
```

```python
#annot
df=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(25,20))
sns.heatmap(df,cmap='hot',annot=True,linewidths=0.5)

#
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]],cmap='Blues')
#
```
