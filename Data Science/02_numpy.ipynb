Broadcasting
It decribe how numpy array deal with different shape while arthmetic operation

```python
#same shape
a=np.arange(6).reshape(2,3)
b=np.arange(6,12).reshape(2,3)
print(a)
print(b)
print(a+b)
```

```python
#diff shape
a=np.arange(6).reshape(2,3)
b=np.arange(3).reshape(1,3)
print(a)
print(b)
print(a+b)
```

Rule of broadcasting
#1.Make two arrays have same number of dimensions 
#  adding 1 in the beginning of the shape of smaller array
#2. Make each dimension of the two arrays same size
#  by repeating the smaller array along the dimension with size 1
#3. Now the two arrays have same shape, we can perform elementwise operation on them 

```python
a=np.arange(12).reshape(2,6)
b=np.arange(9).reshape(3,3)
print(a)
print(b)
#print(a+b) Not possible due to diff shape( intial 1 is not present in b) 
```

```python
a=np.arange(12).reshape(2,6)
b=np.arange(9).reshape(3,3)
print(a)
print(b)
```
