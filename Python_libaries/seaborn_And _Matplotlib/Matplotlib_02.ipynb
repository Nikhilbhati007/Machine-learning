# 3D SURFACE PLOT

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure()
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
plt.colorbar(p)
```

```python
z=np.sin(xx) +np.cos(yy)
fig=plt.figure()
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
plt.colorbar(p)
```

```python
z=np.sqrt(xx**2+yy**2)
fig=plt.figure()
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
plt.colorbar(p)
```

```python
r = 10 # Radius of the sphere
z = np.sqrt(r**2 - xx**2 - yy**2)

fig=plt.figure()
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='plasma')
plt.colorbar(p)
plt.title('3D Surface Plot of a Sphere (Upper Hemisphere)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

```python
#contour plot 3D into 2D
fig=plt.figure()
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis')
plt.colorbar(p)
```

```python
#contour f
fig=plt.figure()
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
plt.colorbar(p)
```
