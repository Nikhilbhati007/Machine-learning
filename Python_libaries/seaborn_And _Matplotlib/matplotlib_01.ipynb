**2D Line plot**

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data






```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
plt.show()
```

```python
#pandas df
batsman=pd.read_csv('/content/sharma-kohli.csv')
batsman.head()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.show()
```

```python
#Multiple plot
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.show()
```

```python
from matplotlib import colors
#label title
plt.plot(batsman['index'],batsman['V Kohli'],color='Black',linestyle='dotted')#dashdot
plt.plot(batsman['index'],batsman['RG Sharma'],color='Red',linestyle='solid',linewidth=3)#dashed
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.show()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],marker='o',color='Black')#+(Marker)
plt.plot(batsman['index'],batsman['RG Sharma'],marker='o',color='Red')#d
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.show()
```

```python
#legends -> location
plt.plot(batsman['index'],batsman['V Kohli'],label='V kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],label='RG Sharma')
plt.legend(loc='upper left')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.show()
```

```python
#limit axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(20000,60000)
plt.xlim(2015,2021)
plt.show()
```

```python
#grids
plt.plot(batsman['index'],batsman['V Kohli'],label='V kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],label='RG Sharma')
plt.legend(loc='upper left')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.grid()
plt.show()
```

```python
#show
plt.plot(batsman['index'],batsman['V Kohli'],label='V kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],label='RG Sharma')
plt.legend(loc='upper left')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.show()
```

#BAR GRAPH
- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
#simple bar chart
child=[10,20,5,40,70]
age_group=['0-2','2-4','4-6','6-8','8-10']
plt.bar(age_group,child)
plt.show
```

```python
#all funct are avilable
#horizontal
plt.barh(age_group,child)
plt.title('Childbirth rate')
plt.xlabel('Age Group')
plt.ylabel('Birth Rate')
plt.show()
```

```python
#multiple bar charts
df=pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')# to avoid overlaping of name
```

```python
#stacked bar chart
plt.bar(df['batsman'],df['2015'],width=0.5)
plt.bar(df['batsman'],df['2016'],width=0.5,bottom=df['2015'])
plt.bar(df['batsman'],df['2017'],width=0.5,bottom=df['2015']+df['2016'])

plt.show()
```

# PIE
- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
